```python
import os

# set the environment variable to control the number of threads
original_omp_num_threads = os.environ.get('OMP_NUM_THREADS', None)
os.environ['OMP_NUM_THREADS'] = '1'

import pyccl as ccl
import numpy as np
from time import time
from multiprocessing import cpu_count
import parallelbar

# generate N random cosmological parameters
rng = np.random.RandomState(14)

n_samples = 1000
Omega_b_arr = rng.uniform(0.04, 0.06, n_samples)
Omega_c_arr = rng.uniform(0.25, 0.35, n_samples)
h_arr = rng.uniform(0.6, 0.8, n_samples)
sigma8_arr = rng.uniform(0.7, 0.9, n_samples)
n_s_arr = rng.uniform(0.9, 1.1, n_samples)

# Generate random n(z)
z = np.linspace(0., 2., 256)
pz = np.exp(-0.5*(z-0.5)**2/0.1**2)

ell = np.arange(2, 2000)

# wrap the parameters and redshift into a list
args = [(Omega_b_arr[i], Omega_c_arr[i], h_arr[i], sigma8_arr[i], n_s_arr[i], z, pz, ell) for i in range(n_samples)]

def wrap_for_parallel(Omega_b, Omega_c, h, sigma8, n_s, z, pz, ell):
    
    cosmo = ccl.Cosmology(Omega_c=Omega_c, Omega_b=Omega_b, h=h, sigma8=sigma8, n_s=n_s)
    # create a tracer for the galaxy distribution
    shearTracer = ccl.WeakLensingTracer(cosmo=cosmo, dndz=(z, pz))

    return ccl.angular_cl(cosmo, shearTracer, shearTracer, ell)

# Now run the function in parallel
n_processes = cpu_count()
print("Using", n_processes, "processes")
start = time()

cls_parallel = list(parallelbar.progress_starmap(wrap_for_parallel, args, n_cpu=n_processes))

# # reset the environment variable
# if original_omp_num_threads is None:
#     del os.environ['OMP_NUM_THREADS']
# else:
#     os.environ['OMP_NUM_THREADS'] = original_omp_num_threads

print("Parallel took", time()-start, "seconds")
```

```python
import matplotlib.pyplot as plt

for i in range(10):
    plt.plot(ell, cls_parallel[i], alpha=0.4)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$C_\ell$')
plt.xscale('log')
plt.yscale('log')
```
